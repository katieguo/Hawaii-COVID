```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import matplotlib.dates as mdates

df = pd.read_csv('https://bitbucket.org/katieguo/katie-test/raw/d974b7b2c7c8471af64e181bc8b0acaf9a35e703/Hawaii_and_other_states/dailynewcases.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')

df2 = pd.read_csv('https://bitbucket.org/katieguo/katie-test/raw/d974b7b2c7c8471af64e181bc8b0acaf9a35e703/Hawaii_and_other_states/dailytests_HawaiiAndOthers.csv')
df2['Date'] = pd.to_datetime(df2['Date'], format='%Y-%m-%d')
      
plt.figure(1, figsize=(12,8))
fig1, ax = plt.subplots(figsize=(12,8))

HI = df.Hawaii
AK = df.Alaska
CA = df.California
CO = df.Colorado
GA = df.Georgia
MT = df.Montana
TX = df.Texas

HI1 = df.Hawaii / df2.Hawaii
AK1 = df.Alaska / df2.Alaska
CA1 = df.California / df2.California
CO1 = df.Colorado / df2.Colorado
GA1 = df.Georgia
MT1 = df.Montana / df2.Montana
TX1 = df.Texas /df2.Texas

#plt.plot(df.Date, HI1, label="Hawaii, SIP: 3/25-5/31")
plt.plot(df.Date, HI1*50, label="Hawaii, SIP: 3/25-5/31")

#plt.plot(df.Date, AK1, label="Alaska, SIP: 3/28-4/24")
#plt.plot(df.Date, CA1, label="California, SIP: 3/19-5/4")
plt.plot(df.Date, CO1, label="Colorado, SIP: 3/26-6/12")
#plt.plot(df.Date, GA1, label="Georgia, SIP: 3/23-4/30")
plt.plot(df.Date, MT1, label="Montana, SIP: 3/27-4/26")
plt.plot(df.Date, TX1, label="Texas, SIP: 4/2-4/30")

plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  


plt.title('COVID-19 Daily New Cases / Daily Tested in Hawaii & Other States')
plt.ylabel('New Daily Cases / Daily New Tests')
plt.legend()

      
plt.figure(2, figsize=(12,8))
fig1, ax = plt.subplots(figsize=(12,8))

HI2 = HI1.rolling(7).mean()
AK2 = AK1.rolling(7).mean()
CA2 = CA1.rolling(7).mean()
CO2 = CO1.rolling(7).mean()
#GA2 = GA1.rolling(7).mean()
MT2 = MT1.rolling(7).mean()
TX2 = TX1.rolling(7).mean()

plt.plot(df.Date, HI2, label="Hawaii, SIP: 3/25-5/31")
#plt.plot(df.Date, AK2, label="Alaska, SIP: 3/28-4/24")
#plt.plot(df.Date, CA2, label="California, SIP: 3/19-5/4")
plt.plot(df.Date, CO2, label="Colorado, SIP: 3/26-6/12")
#plt.plot(df.Date, GA2, label="Georgia, SIP: 3/23-4/30")
plt.plot(df.Date, MT2, label="Montana, SIP: 3/27-4/26")
plt.plot(df.Date, TX2, label="Texas, SIP: 4/2-4/30")

plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  


plt.title('COVID-19 Daily New Cases / Daily Tested in Hawaii & Other States, 7-day rolling average')
plt.ylabel('New Daily Cases / Daily New Tests')
plt.legend()



plt.figure(3, figsize=(12,8))
fig1, ax = plt.subplots(figsize=(12,8))

HI3 = HI.cumsum() / df2.Hawaii.cumsum()
AK3 = AK.cumsum() / df2.Alaska.cumsum()
CA3 = CA.cumsum() / df2.California.cumsum()
CO3 = CO.cumsum() / df2.Colorado.cumsum()
#GA3 = GA.cumsum() / df2.Georgia.cumsum()
MT3 = MT.cumsum() / df2.Montana.cumsum()
TX3 = TX.cumsum() / df2.Texas.cumsum()

plt.plot(df.Date, HI3, label="Hawaii, SIP: 3/25-5/31")
plt.plot(df.Date, AK3, label="Alaska, SIP: 3/28-4/24")
plt.plot(df.Date, CA3, label="California, SIP: 3/19-5/4")
plt.plot(df.Date, CO3, label="Colorado, SIP: 3/26-6/12")
#plt.plot(df.Date, GA3, label="Georgia, SIP: 3/23-4/30")
plt.plot(df.Date, MT3, label="Montana, SIP: 3/27-4/26")
plt.plot(df.Date, TX3, label="Texas, SIP: 4/2-4/30")

plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  


plt.title('COVID-19 Cumulative Cases / Cumulative Tested in Hawaii & Other States, Cumulative Sum')
plt.ylabel('Cumulative Cases / Cumulative Tested')
plt.legend()



plt.figure(4, figsize=(12,8))
fig1, ax = plt.subplots(figsize=(12,8))

HI4 = HI3.rolling(7).mean()
AK4 = AK3.rolling(7).mean()
CA4 = CA3.rolling(7).mean()
CO4 = CO3.rolling(7).mean()
#GA4 = GA3.rolling(7).mean()
MT4 = MT3.rolling(7).mean()
TX4 = TX3.rolling(7).mean()

plt.plot(df.Date, HI4, label="Hawaii, SIP: 3/25-5/31")
plt.plot(df.Date, AK4, label="Alaska, SIP: 3/28-4/24")
plt.plot(df.Date, CA4, label="California, SIP: 3/19-5/4")
plt.plot(df.Date, CO4, label="Colorado, SIP: 3/26-6/12")
#plt.plot(df.Date, GA4, label="Georgia, SIP: 3/23-4/30")
plt.plot(df.Date, MT4, label="Montana, SIP: 3/27-4/26")
plt.plot(df.Date, TX4, label="Texas, SIP: 4/2-4/30")

plt.xticks(rotation=45)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))  


plt.title('COVID-19 Cumulative Cases / Cumulative Tested in Hawaii & Other States, Cumulative Sum 7-day Rolling Average')
plt.ylabel('Cumulative Cases / Cumulative Tests')
plt.legend()


```
