```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df1 = pd.read_csv('hawaii.csv', delimiter=' ')

Hawaii_month=df1.month
Hawaii_daily=df1.daily

print(df1)
#ax=df1.plot(kind='scatter',x='Hawaii_month',y='Hawaii_daily',color='red',marker='o',figsize=(18,12))
plt.figure(figsize=(12,8))
plt.scatter(Hawaii_month,Hawaii_daily,marker='o',facecolors='none', edgecolors='r')
ax=plt.gca()

df2 = pd.read_csv('US.csv', delimiter=' ')
print(df2)
US_month=df2.month
US_daily=df2.daily
US_daily2=US_daily*1000/251.25

ax.scatter(US_month,US_daily,marker='s', facecolors='none', edgecolors='b')
ax.scatter(US_month,US_daily2,marker='d', facecolors='none', edgecolors='g')
plt.title('COVID-19 cases in Hawaii & US 2020')
plt.xlabel('month + day/(days in month)')
plt.ylabel('daily cases HI (o) or US/1000 (square), US/251.25 (diamond)')
plt.xticks([4,5,6,7,8])
plt.yscale('log')
#plt.yticks([0,1,2,5,10,20,50])
```
