```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.dates as mdates
from datetime import datetime

# A list of COVID events and their dates
names = ['Apr 1917: U.S. enters WW1 with 378,000 in the armed services',
         'Jun 1917: A military draft is established;\nArmy begins training recruits at 32 large camps',
         'Mar 1918: Outbreaks of flu-like illness are first detected in the U.S.;\nMore than 100 soldiers at Camp Funston in Kansas become ill with flu;\nSporadic flu activity spreads unevenly through the U.S.,\nEurope, and possibly Asia over the next six months',
         'Apr 1918: First mention of influenza appears in an April 5\nweekly public health report.18 severe cases and 3 deaths in Haskell, Kansas',
         'May 1918: Hundreds of thousands of soldiers are\ndeployed across the Atlantic each month',
         'Sep 1918: Second wave of influenza emerges in Boston;\nBetween September and November, a second wave of flu peaks in the U.S. and is responsible\nfor most of the deaths attributed to the pandemic. NYC\'s Board of Health adds flu to the list of reportable diseases\nand requires all flu cases to be isolated at home or in a city hospital',
         'Oct 1918: 195,000 Americans die of flu in October;\nThe U.S. experiences severe shortages of professional nurses; Chicago, alongside others in the\nU.S., closes theaters,movie houses, night schools, and prohibits public gatherings;\nSan Francisco\'s Board of Health requires any person serving the public to wear masks\nand issues strong recommendation to all residents to wear masks in public;\nNew York City reports a 40% decline in shipyard productivity',
         'Nov 1918: End of WW1 enables a resurgence of influenza as people\ncelebrate Armistice Day and soldiers begin to demobilize;\nSalt Lake City officials place quarantine signs on doors of 2,000 homes',
         'Dec 1918: Public health officials begin education programs\nand publicity about dangers of coughing and sneezing;\nCommitte of the American Public Health Association encourages stores and factories to stagger opening and closing hours\nand for people to walk to work to prevent overcrowding on public transport',
         'Jan 1919: Third wave of influenza occurs in winter and spring of 1919, subsides in the summer;\n1,800 flu cases and 101 deaths in San Francisco in the first five days of January;\nSan Antonio citizens complain that new flu cases are not being reported,\nwhich is fueling another influenza surge;\n767 flu cases and 67 deaths are reported in New York City;\nTrustees of the Boston City Hospital ask mayor for a special appropriation\nof $3,000 to study the treatment of influenza',
         'Feb 1919: Influenza appears to be nearly eradicated in New Orlands;\nIllinois passes a bill to create a one-year course to become a "practical\nnurse", an effort to address the nursing shortage that the pandemic had exposed',
         'Apr 1919: U.S. President Woodrow Wilson collapses at Versailles Peace\nConference while negotiating the end of WW1. Some historians speculate\nthat he was weak from influenza, which was still rampant in Paris']

dates = ['1917-04', '1917-06', '1918-03', '1918-04', '1918-05', '1918-09',
         '1918-10', '1918-11', '1918-12', '1919-01', '1919-02', '1919-04']
dates = [datetime.strptime(ii, "%Y-%m") for ii in dates]

# Choose some nice levels
levels = np.tile([-1, 1, -4, 4, -7, 7],
                 int(np.ceil(len(dates)/6)))[:len(dates)]

# Create figure and plot a stem plot with the date
fig, ax = plt.subplots(figsize=(24, 10), constrained_layout=True)
plt.suptitle('Influenza (Flu) Pandemic Events and Dates in 1917-1919', fontsize=24)
markerline, stemline, baseline = ax.stem(dates, levels,
                                         linefmt="C3-", basefmt="k-",
                                         use_line_collection=True)

plt.setp(markerline, mec="k", mfc="w", zorder=3)

# Shift the markers to the baseline by replacing the y-data by zeros.
markerline.set_ydata(np.zeros(len(dates)))

# annotate lines
vert = np.array(['top', 'bottom'])[(levels > 0).astype(int)]
for d, l, r, va in zip(dates, levels, names, vert):
    ax.annotate(r, xy=(d, l), xytext=(-3, np.sign(l)*3),
                textcoords="offset points", va=va, ha="center")

# format xaxis with 7 day intervals
ax.get_xaxis().set_major_locator(mdates.MonthLocator(interval=1))
ax.get_xaxis().set_major_formatter(mdates.DateFormatter("%b %Y"))
plt.setp(ax.get_xticklabels(), rotation=30, ha="right")

# remove y axis and spines
ax.get_yaxis().set_visible(False)
for spine in ["left", "top", "right"]:
    ax.spines[spine].set_visible(False)

ax.margins(y=0.1)
plt.show()
```
